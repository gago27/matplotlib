```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
%matplotlib inline
archivo = pd.read_csv('indicadores-mundiales.csv',
                     sep = ',',
                     encoding = 'latin-1')
archivo['Region'] = archivo['Region'].map(lambda x: x.capitalize())
dato = archivo.groupby('Region').sum()[['Poblacion 0-14',
       'Poblacion 15-64', 'Poblacion 65+']]
datos = dato.T


fig, axs = plt.subplots(1, 6, sharex = 'col', sharey = 'row', figsize =(17,9));
plt.suptitle('Region')
axs[0].set_ylabel('Valor')
plt.style.use('classic')
plt.subplots_adjust(wspace = 0.005, hspace=0.01)
color = ['blue', 'orange', 'green']
for index, colname in enumerate(datos.columns):
  axs[index].scatter(list(datos.index), datos[colname]/100, s = 300, edgecolors = color, c='white', linewidths = 3)
  axs[index].set_xlim(-0.3, 2.3)
  axs[index].set_xticklabels(datos.index, rotation='vertical')
  axs[index].set_title(colname)
```
