```
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

archivo = pd.read_csv('indicadores-mundiales.csv',
                     sep= ',',
                     encoding = 'latin-1')
archivo['Region'] = archivo['Region'].map(lambda x: x.capitalize())
datos = archivo.groupby('Region').sum()[['Poblacion 0-14', 'Poblacion 15-64', 'Poblacion 65+']]
dato = datos.T

fig, axs = plt.subplots(3, 1, sharex = 'none', sharey = 'row', figsize= (8, 15),)
plt.subplots_adjust(hspace = 0.03, )
color = ['blue', 'orange', 'green']
for index, colname in enumerate(datos.columns):
  axs[index].bar(list(datos.index), datos[colname]/100, color = color[index])
  axs[index].legend(datos.columns)
  axs[index].set_ylabel('Valor')
  
```
